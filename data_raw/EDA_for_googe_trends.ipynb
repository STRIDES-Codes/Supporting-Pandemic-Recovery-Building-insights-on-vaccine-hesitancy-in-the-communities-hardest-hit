# Google Trends Data
<hr>

```python
import pandas as pd 
import numpy as np
import os 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../data_raw/Google_Trends/geoMap (14).csv', skiprows = 1)
df2 = pd.read_csv('../data_raw/Google_Trends/geoMap (15).csv', skiprows = 1)
df3 = pd.read_csv('../data_raw/Google_Trends/geoMap (16).csv', skiprows = 1)
df4 = pd.read_csv('../data_raw/Google_Trends/geoMap (17).csv', skiprows = 1)
df5 = pd.read_csv('../data_raw/Google_Trends/geoMap (18).csv', skiprows = 1)
df6 = pd.read_csv('../data_raw/Google_Trends/geoMap (19).csv', skiprows = 1)
df7 = pd.read_csv('../data_raw/Google_Trends/geoMap (20).csv', skiprows = 1)
df8 = pd.read_csv('../data_raw/Google_Trends/geoMap (21).csv', skiprows = 1)
```

```python
df1.head()
```

# Merge
<hr>

```python
master_df = df1.merge(df2, on = 'Region', how = 'inner')
master_df = master_df.merge(df3, on = 'Region', how = 'inner')
master_df = master_df.merge(df4, on = 'Region', how = 'inner')
master_df = master_df.merge(df5, on = 'Region', how = 'inner')
master_df = master_df.merge(df6, on = 'Region', how = 'inner')
master_df = master_df.merge(df7, on = 'Region', how = 'inner')
master_df = master_df.merge(df8, on = 'Region', how = 'inner')
```

```python
master_df.head()
```

```python
# fill in NaN to 1's
master_df = master_df.fillna(value = 1)
```

# EDA
